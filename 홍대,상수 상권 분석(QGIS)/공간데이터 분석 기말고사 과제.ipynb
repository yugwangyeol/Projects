```python
import pandas as pd
```

# Data Load

```python
df_2010 =  pd.read_csv('2010.csv',encoding='cp949')
df_2020 =  pd.read_csv('2020.csv',encoding='cp949')
```

```python
print(df_2010.shape)
display(df_2010.head())
```

```python
print(df_2020.shape)
display(df_2020.head())
```

```python
intersection = list(set(df_2010.STORE_NM) & set(df_2020.상호명))
```

```python
df_10 = df_2020.loc[df_2020['상호명'].isin(intersection)].drop_duplicates(['상호명']).reset_index(drop=True)
display(df_10.head())
print(df_10.shape)
```

```python
df_10.상권업종_2.unique()
```

```python
df_10_l = ['양식','제과제빵떡케익', '커피점/카페', '한식']
df_2020 = df_10.loc[df_10['상권업종_2'].isin(df_10_l)].reset_index(drop=True)
display(df_2020)
print(df_2020.shape)
```

```python
df_10 = df_2010.loc[df_2010['STORE_NM'].isin(intersection)].drop_duplicates(['STORE_NM']).reset_index(drop=True)
display(df_10.head())
print(df_10.shape)
```

```python
df_10.UPJONG2_NM.unique()
```

```python
df_10_l = [ '다방/커피숍/카페', '한식', '제과제빵떡케익', '양식', '분식']
df_2010 = df_10.loc[df_10['UPJONG2_NM'].isin(df_10_l)].reset_index(drop=True)
display(df_2010)
print(df_2010.shape)
```
