```python
import pandas as pd
```

```python
a=pd.read_csv('강남구_대치_ 2016.csv',encoding='cp949')
b=pd.read_csv('강서구_화곡동_2017.csv',encoding='cp949')
c=pd.read_csv('강서구_화곡동_2018.csv',encoding='cp949')
d=pd.read_excel('광진 구의동 2019.xls',encoding='cp949')
e=pd.read_excel('광진 구의동 2020.xls',encoding='cp949')
f=pd.read_csv('광진구_구의동_2015.csv',encoding='cp949')
g=pd.read_csv('광진구_구의동_2016.csv',encoding='cp949')
h=pd.read_csv('광진구_구의동_2017.csv',encoding='cp949')
i=pd.read_csv('광진구_구의동_2018.csv',encoding='cp949')
j=pd.read_csv('서초구_서초동_2013.csv',encoding='cp949')
```

```python
a = a.iloc[1:,:]
b = b.iloc[1:,:]
c = c.iloc[1:,:]
d = d.iloc[1:,:]
e = e.iloc[1:,:]
f = f.iloc[1:,:]
g = g.iloc[1:,:]
h = h.iloc[1:,:]
i = i.iloc[1:,:]
j = j.iloc[1:,:]
```

```python
df = pd.concat([a,b,c,d,e,f,g,h,i,j])
df = df.ffill().reset_index()
```

```python
g = pd.DataFrame(df.계약일.str.split('\n\r'))
h = pd.DataFrame(df.거래금액.str.split('\n\r'))
j = pd.DataFrame(df.층.str.split('\n\r'))
```

```python
g = g['계약일'].apply(pd.Series).reset_index().melt(id_vars='index').dropna()[['index', 'value']].set_index('index')

h = h['거래금액'].apply(pd.Series).reset_index().melt(id_vars='index').dropna()[['index', 'value']].set_index('index')

j = j['층'].apply(pd.Series).reset_index().melt(id_vars='index').dropna()[['index', 'value']].set_index('index')
```

```python
t = pd.concat([g,h],axis=1)
w = pd.concat([t,j],axis=1)
w
```

```python
w = w.dropna(axis=0)
w.columns = ['계약일','거래금액','층']
w
```

```python
w = w[w.층 != '']
w
```

```python
k = pd.DataFrame(df['단지[준공년도]'].str.split('\n'))
l = k['단지[준공년도]'].apply(pd.Series).reset_index().melt(id_vars='index').dropna()[['index', 'value']].set_index('index')
l

ke = l.iloc[:int(len(l)/2),:]
kl = l.iloc[int(len(l)/2):,:]
```

```python
df_s = pd.merge(w,ke,left_index=True,right_index=True,how='left')
df_s
```

```python
df = df[['단지[준공년도]','건물전용면적(㎡)','계약년도']]
df['단지[준공년도]'] = ke
df
```

```python
df_ss = pd.merge(df,w,left_index=True, right_index=True, how='left')
df_ss = pd.merge(df_ss,kl,left_index=True, right_index=True, how='left')
df_ss = df_ss.rename(columns = {'value':'준공년도'})
df_ss
```

```python
def gh(x):
    return x.strip('[]')

df_ss['준공년도'] = df_ss['준공년도'].apply(gh).astype(int)
df_ss['계약년도'] = df_ss['계약년도'].astype(int)
df_ss['건물연령'] = df_ss['계약년도'] - df_ss['준공년도']
df_ss
```

```python
df_ss = df_ss.rename(columns = {'단지[준공년도]':'단지','계약일':'계약일','건물전용면적(㎡)':'건물전용면적'})
df_ss
```

```python
df_ss.to_csv('예주.csv',index=False)
```
