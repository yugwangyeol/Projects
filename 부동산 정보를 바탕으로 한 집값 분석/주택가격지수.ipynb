```python
import pandas as pd
```

```python
df = pd.read_excel('월간_매매가격지수_아파트.xlsx')
df.head(10)
```

```python
df['지역'] = df['지역'].ffill()
df
```

```python
df_1 = df.query('지역 == "서울"')
df_1
```

```python
df_2 = df_1.drop([df_1.index[0],df_1.index[1],df_1.index[16]])
df_2
```

```python
df_3 = df_2.drop(df.columns[[0,1,2]], axis='columns')
df_3
```

```python
a = []

for i in range(2010,2021):
    for j in range(1,13):
        if i == 2020:
            if j == 7:
                break
            else:
                a.append(str(i)+'년 '+str(j)+'월_지수')
                a.append(str(i)+'년 '+str(j)+'월_변동')
        else:
            a.append(str(i)+'년 '+str(j)+'월_지수')
            a.append(str(i)+'년 '+str(j)+'월_변동')
            
a.insert(0,'지역')
a
```

```python
df_3.columns = a
```

```python
colname = df_3.columns.tolist()

for i in range(1,253,2):
    if i == 1:
        주택가격지수 = pd.concat([df_3['지역'],df_3.iloc[:,i:i+2]],axis = 1)
        주택가격지수.rename(columns = {주택가격지수.columns[1]:'지수',주택가격지수.columns[2]:'변동'}, inplace = True)
        주택가격지수['연도'] = colname[i][:4]
        주택가격지수['월'] = colname[i][6:-4]
    else:
        #주택가격지수 = pd.concat([주택가격지수,pd.concat([df_3['지역'],df_3.iloc[:,i:i+2]],axis = 1)])
        x = pd.concat([df_3['지역'],df_3.iloc[:,i:i+2]],axis = 1)
        x.rename(columns = {x.columns[1]:'지수',x.columns[2]:'변동'}, inplace = True)
        x['연도'] = colname[i][:4]
        x['월'] = colname[i][6:-4]
        주택가격지수 = pd.concat([주택가격지수,x])
```

```python
주택가격지수 = 주택가격지수.reset_index()
del 주택가격지수['index']
주택가격지수
```

```python
주택가격지수.to_csv('주택가격지수.csv',index=False)
```
