```python
# 사용하는 모듈 불러오기
from selenium import webdriver
import time
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.alert import Alert

# csv 파일 만들기
f = open('seoul_apart.csv','w')
f.write("단지,준공년도,전용면적,계약년도,계약일,거래금액,층,자치구,법정동\n")

# 드라이버 실행하기
driver = webdriver.Chrome('./chromedriver')
driver.get('http://land.seoul.go.kr/land/rtms/rtmsApartment.do')

# Alert 제어하기
da = Alert(driver)

# 분기 -> 기간
peroid_xpath = Select(driver.find_element_by_xpath('/html/body/div[1]/div/div/div[3]/div[1]/form/dl[2]/dd/div[1]/select[1]'))
peroid_xpath.select_by_visible_text('기간')
time.sleep(0.3)

# 구 리스트 만들기
gu_xpath = driver.find_element_by_xpath('/html/body/div[1]/div/div/div[3]/div[1]/form/dl[1]/dd/div[1]/select[2]')
gu_list = gu_xpath.text.split('\n')[1:]
time.sleep(0.3)

# 구 선택하기
for gu in gu_list:
    Select(gu_xpath).select_by_visible_text(gu)
    time.sleep(0.3)
    
    # 동 리스트 만들기
    dong_xpath = driver.find_element_by_xpath('/html/body/div[1]/div/div/div[3]/div[1]/form/dl[1]/dd/div[2]/select[3]')
    dong_list = dong_xpath.text.split('\n')[1:]
    
    # 동 선택하기
    for dong in dong_list:
        Select(dong_xpath).select_by_visible_text(dong)
        time.sleep(0.3)
        
        year2 = [i for i in range(6,16)]
        day2 = [7,1,3,4,5,6,1,2,3]

        for i in range(0,10):
            if i < 9:
                # 기간 지정하기 - 2010 ~ 2019
                # 캘린더2 클릭하기
                calendar2_button = driver.find_element_by_css_selector('div#select2.hide img.ui-datepicker-trigger:nth-of-type(2)')
                calendar2_button.click()
                time.sleep(0.3)
                
                # 년 선택하기
                year2_button = driver.find_element_by_css_selector(f'select.ui-datepicker-year option:nth-of-type({year2[i]})')
                year2_button.click()
                time.sleep(0.3)
                
                # 월 선택하기
                month2_button = driver.find_element_by_css_selector('select.ui-datepicker-month > option:nth-of-type(1)')
                month2_button.click()
                time.sleep(0.3)
                
                # 일 선택하기
                day2_botton = driver.find_element_by_css_selector(f'table.ui-datepicker-calendar tbody tr:nth-of-type(1) td:nth-of-type({day2[i]})')
                day2_botton.click()
                time.sleep(0.3)
                da.accept()
                
                #검색 버튼 누르기
                search_button = driver.find_element_by_css_selector('input#search.on')
                search_button.click()
                time.sleep(3)
                
                # 컨테이너 : tbody#resultList2 tr
                data = driver.find_elements_by_css_selector('tbody#resultList2 tr')

                for d in data:
                    # 데이터가 없는 경우 NaN값으로 채우기
                    try:
                        단지_준공년도 = d.find_element_by_css_selector('td.tb_bottom:nth-of-type(1)').text
                        단지 = 단지_준공년도[:-7]
                        준공년도 = 단지_준공년도[-5:-1]
                        전용면적 = d.find_element_by_css_selector('td.tb_bottom:nth-of-type(3)').text
                        거래년도 = d.find_element_by_css_selector('td.tb_top.tb_bottom:nth-of-type(4)').text
                        거래일 = d.find_element_by_css_selector('td.tb_top.tb_bottom:nth-of-type(5)').text
                        거래금액 = d.find_element_by_css_selector('td.tb_top.tb_bottom:nth-of-type(6)').text
                        층 = d.find_element_by_css_selector('td.tb_top.tb_bottom:nth-of-type(7)').text
                    except:
                        단지 = 'NaN'
                        준공년도 = 'NaN'
                        전용면적 = 'NaN'
                        거래년도 = 'NaN'
                        거래일 = 'NaN'
                        거래금액 = 'NaN'
                        층 = 'NaN'
                    
                    거래금액 = 거래금액.replace(',','')
                    
                    print('단지 :',단지)
                    print('준공년도 :',준공년도)
                    print('전용면적 :',전용면적)
                    print('거래년도 :',거래년도)
                    print('거래일 :',거래일)
                    print('거래금액 :',거래금액)
                    print('층 :',층)
                    print('자치구 :',gu)
                    print('동 : ',dong)
                    print('=======================')
                    
                    f.write(단지+ "," +준공년도+ "," +전용면적+ "," +거래년도+ "," +거래일+ "," +거래금액+ "," +층+ "," +gu+ "," +dong+ "\n")
            else:
                # 기간 지정하기 - 2020
                # 캘린더1 클릭하기
                calendar1_button = driver.find_element_by_css_selector('div#select2.hide img.ui-datepicker-trigger:nth-of-type(1)')
                calendar1_button.click()
                time.sleep(0.3)
                
                # 년 선택하기
                year1_button = driver.find_element_by_css_selector('select.ui-datepicker-year option:nth-of-type(15)')
                year1_button.click()
                time.sleep(0.3)
                
                # 월 선택하기
                month1_button = driver.find_element_by_css_selector('select.ui-datepicker-month > option:nth-of-type(1)')
                month1_button.click()
                time.sleep(0.3)
                
                # 일 선택하기
                day1_button = driver.find_element_by_css_selector('table.ui-datepicker-calendar tbody tr:nth-of-type(1) td:nth-of-type(4)')
                day1_button.click()
                time.sleep(0.3)
                da.accept()

                # 캘린더2 클릭하기
                calendar2_button = driver.find_element_by_css_selector('div#select2.hide img.ui-datepicker-trigger:nth-of-type(2)')
                calendar2_button.click()
                time.sleep(0.3)
                
                # 년 선택하기
                year2_button = driver.find_element_by_css_selector(f'select.ui-datepicker-year option:nth-of-type(15)')
                year2_button.click()
                time.sleep(0.3)
                
                # 월 선택하기
                month2_button = driver.find_element_by_css_selector('select.ui-datepicker-month > option:nth-of-type(6)')
                month2_button.click()
                time.sleep(0.3)
                
                # 일 선택하기
                day2_botton = driver.find_element_by_css_selector('table.ui-datepicker-calendar tbody tr:nth-of-type(5) td:nth-of-type(3)')
                day2_botton.click()
                time.sleep(0.3)

                #검색 버튼 누르기
                search_button = driver.find_element_by_css_selector('input#search.on')
                search_button.click()
                time.sleep(5)
        
                data = driver.find_elements_by_css_selector('tbody#resultList2 tr')

                for d in data:
                    # 데이터가 없는 경우 NaN값으로 채우기
                    try:
                        단지_준공년도 = d.find_element_by_css_selector('td.tb_bottom:nth-of-type(1)').text
                        단지 = 단지_준공년도[:-7]
                        준공년도 = 단지_준공년도[-5:-1]
                        전용면적 = d.find_element_by_css_selector('td.tb_bottom:nth-of-type(3)').text
                        거래년도 = d.find_element_by_css_selector('td.tb_top.tb_bottom:nth-of-type(4)').text
                        거래일 = d.find_element_by_css_selector('td.tb_top.tb_bottom:nth-of-type(5)').text
                        거래금액 = d.find_element_by_css_selector('td.tb_top.tb_bottom:nth-of-type(6)').text
                        층 = d.find_element_by_css_selector('td.tb_top.tb_bottom:nth-of-type(7)').text
                    except:
                        단지 = 'NaN'
                        준공년도 = 'NaN'
                        전용면적 = 'NaN'
                        거래년도 = 'NaN'
                        거래일 = 'NaN'
                        거래금액 = 'NaN'
                        층 = 'NaN'
                    
                    거래금액 = 거래금액.replace(',','')
                    
                    print('단지 :',단지)
                    print('준공년도 :',준공년도)
                    print('전용면적 :',전용면적)
                    print('거래년도 :',거래년도)
                    print('거래일 :',거래일)
                    print('거래금액 :',거래금액)
                    print('층 :',층)
                    print('자치구 :',gu)
                    print('동 : ',dong)
                    print('=======================')
                    
                    f.write(단지+ "," +준공년도+ "," +전용면적+ "," +거래년도+ "," +거래일+ "," +거래금액+ "," +층+ "," +gu+ "," +dong+ "\n")

f.close()
driver.close()
```
