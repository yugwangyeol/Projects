# 통합데이터 편집

```python
df_1['건물연령'] = df_1['계약년도']-df_1['준공년도']
df_1
```

```python
df_1['계약일'] = df_1['계약일'].apply(lambda x: int(x)) 
df_1
```

```python
def dd(x):
    if x == 1 or x == 2 or x == 3:
        return 1
    elif x == 4 or x == 5 or x == 6:
        return 2
    elif x == 7 or x == 8 or x == 9:
        return 3
    else:
        return 4

df_1['계약일'] = df_1.계약일.apply( dd )
df_1
```

```python
df_1.rename(columns={'계약일':'분기'},inplace=True)
df_1
```

```python
df_1 = df_1.groupby(['행정동','계약년도','분기']).agg({'거래금액':'sum','단지':'count'}).reset_index()
df_1
```

```python
df_1['금액'] = round(df_1['거래금액']/df_1['단지'],0)
df_1
```

```python
동 = df_1.행정동.unique().tolist()
w = []

for i in 동:
    q = df_1.query('행정동 == @i').reset_index()
    del q['index']
    for i in range(0,len(q)):
        if i == 0:
            w.append(0)
        elif i == len(q):
            w.append(0)
        else:
            w.append(round((q.iloc[i,5]/q.iloc[i-1,5])-1,3))
w
```

```python
df_1['상하강률'] = w
df_1
```

```python
df_1.rename(columns={'계약년도':'년'},inplace=True)
df_1
```

```python
dtrain = xgb.DMatrix(x_train, label=y_train)
dtest = xgb.DMatrix(x_test)
```

```python
def xgb_evaluate(max_depth, gamma, colsample_bytree, eta, min_child_weight, subsample):
    params = {'eval_metric': 'rmse',
              'max_depth': int(max_depth),
              'subsample': float(subsample),
              'eta': float(eta),
              'gamma': gamma,
              'colsample_bytree': colsample_bytree,
              'min_child_weight' : int(min_child_weight)}
    # Used around 1000 boosting rounds in the full model
    cv_result = xgb.cv(params, dtrain, num_boost_round=100, nfold=3)    
    
    # Bayesian optimization only knows how to maximize, not minimize, so return the negative RMSE
    return -1. * cv_result['test-rmse-mean'].iloc[-1]
```

```python
xgb_bo = BayesianOptimization(xgb_evaluate, {'max_depth': (2, 5), 
                                             'gamma': (0, 1),
                                             'colsample_bytree': (0.3, 0.9),
                                              'eta' : (.05, .1),
                                             'min_child_weight' : (3,6),
                                             'subsample':(0.6, 0.8)})

# Use the expected improvement acquisition function to handle negative numbers
# Optimally needs quite a few more initiation points and number of iterations
xgb_bo.maximize(init_points=3, n_iter=15, acq='ei')
```

```python
# best parameters
params = xgb_bo.max['params']
params['max_depth'] = int(params['max_depth'])
```

```python
model3 = xgb.train(params, dtrain, num_boost_round=100)
```

```python
result = model3.predict(xgb.DMatrix(x_test))
```
