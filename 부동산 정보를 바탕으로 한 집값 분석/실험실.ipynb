```python
import pandas as pd
```

```python
df = pd.read_csv('통합데이터.csv',encoding='cp949')
df
```

```python
df['계약일'] = df['계약일'].apply(lambda x: int(x)) 
df
```

```python
def dd(x):
    if x == 1 or x == 2 or x == 3:
        return 1
    elif x == 4 or x == 5 or x == 6:
        return 2
    elif x == 7 or x == 8 or x == 9:
        return 3
    else:
        return 4

df['계약일'] = df.계약일.apply( dd )
df
```

```python
df.rename(columns={'계약일': '분기'}, inplace=True)
df
```

```python
df['건물연령'] = df['계약년도']-df['준공년도']
df
```

```python
del df['법정동']
df = df[['자치구','행정동','계약년도','분기','단지','건물연령','거래금액']]
df
```

```python
df.sort_values(by = ['자치구','행정동','계약년도','분기','단지'])
df
```
